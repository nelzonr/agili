```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# importar dataset de posicões gps dos veiculos
p = pd.read_csv('posicoes_um_dia.csv')
p.head()
```

```python
# importar dataset dos pontos de ônibus
pontos_onibus = pd.read_csv('pontos_onibus.txt')
pontos_onibus.head()
```

```python
# transformação, ordenação e exclusão de variáveis
df = p["codigo_linha sentido linha id_veiculo latitude longitude data_hora".split()]
df['codigo_linha_sentido'] = p['codigo_linha'].map(str) + '-' + p['sentido'].map(str)
df['letreiro'] = p['letreiro_ida'] + ' # ' + p['letreiro_volta']
df = df["codigo_linha_sentido linha letreiro id_veiculo latitude longitude data_hora".split()]
```

```python
df.head()
```

```python
df.to_csv("dataframe_posicoes.csv")
```

```python
# carregar e posicionar o mapa no centro de São Paulo
mapa = folium.Map([-23.550169, -46.634091], zoom_start=12)
mapa
```

```python
# mostrar o mapa de calor
posicoesArray = df[['latitude', 'longitude']].values
mapa.add_child(plugins.HeatMap(posicoesArray, radius=14))
mapa
```

```python
# HeatMap na linha N101
linha_n101 = df[df['linha'] == 'N101']
mapa = folium.Map([-23.550169, -46.634091], zoom_start=12)
posicoesArray = linha_n101[['latitude', 'longitude']].values
mapa.add_child(plugins.HeatMap(posicoesArray, radius=14))
mapa
```
